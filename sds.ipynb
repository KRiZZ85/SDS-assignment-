```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
age_mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean)
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Mode_of_Transport'] = df['Mode_of_Transport'].fillna(df['Mode_of_Transport'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Histogram for Age
df['Age'].hist(bins=20, color='skyblue', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Histogram for Work Hours
df['Work_Hours'].hist(bins=20, color='orange', alpha=0.7)
plt.title('Distribution of Work Hours')
plt.xlabel('Work Hours')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Age
sns.boxplot(y=df['Age'], color='lightblue')
plt.title('Boxplot of Age')
plt.show()
```

```python
# Select only numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix
corr_matrix = df[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, square=True)
plt.title('Correlation Matrix for Numerical Features', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
